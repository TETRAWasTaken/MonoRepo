```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

### Numpy Linear Spacing

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
```

### Subplots

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(x, y, 'b-')
```

## Object Oriented Programming with Matplot

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'r--')
axes.set_title('Sine Wave')
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
```

### Nested Plots

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x, y, "--b")
axes1.set_xlabel("X-axis")
axes1.set_ylabel("Y-axis")
axes1.set_title("Sine Wave")

axes2.plot(x, y, "r-")
axes2.set_xlabel("X-axis")
axes2.set_ylabel("Y-axis")
axes2.set_title("Cosine Wave")
```

### Creating Subplot

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for i in axes:
    i.set_xlabel("X-axis")
    i.set_ylabel("Y-axis")
    i.set_title("Sine Wave")
    i.plot(x, y, 'b--')
```

## Saving the Figure

```python
fig.savefig("FILES/my_figure.png", dpi=200)
```

## Adding Legend

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, label='Sine Wave', color='b', linestyle='--', linewidth=2, marker='o', markersize=4, markerfacecolor='r')
ax.plot(x, np.cos(x), label='Cosine Wave')
ax.legend(loc=4)
```

## Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
langs = ['Python', 'Java', 'C++', 'Ruby', 'JavaScript']
students = [22, 14, 17, 10, 19]
ax.bar(langs, students)
plt.show()
```

### Multiple Bar Chart

```python
data = [[5, 25, 50, 20], [4, 23, 51, 17], [6, 22, 52, 19]]
x = np.arange(4)
width = 0.2
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(x - width, data[0], width, label='2015')
ax.bar(x, data[1], width, label='2016')
ax.bar(x + width, data[2], width, label='2017')
ax.set_ylabel('Y-axis')
ax.set_title('Bar Plot')
ax.set_xticks(x)
ax.set_xticklabels(['A', 'B', 'C', 'D'])
ax.legend()
plt.show()
```

## Stack Plot

```python
# Simple stacked bar plot example
labels = ['A', 'B', 'C', 'D']
men = np.array([20, 35, 30, 35])
women = np.array([25, 32, 34, 20])
children = np.array([15, 25, 20, 15])

fig, ax = plt.subplots()
ax.bar(labels, men, label='Men')
ax.bar(labels, women, bottom=men, label='Women')
ax.bar(labels, children, bottom=men+women, label='Children')

ax.set_ylabel('Scores')
ax.set_title('Stacked Bar Plot')
ax.legend()
plt.show()
```

## Histogram

```python
# Simple histogram with bins
np.random.seed(0)
data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with 30 bins')
plt.show()
```

## Grid Plotting

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Top-left plot
axes[0, 0].plot(x, np.sin(x), 'b-')
axes[0, 0].set_title('Sine Wave')

# Top-right plot
axes[0, 1].plot(x, np.cos(x), 'r--')
axes[0, 1].set_title('Cosine Wave')

# Bottom-left plot
axes[1, 0].plot(x, np.tan(x), 'g.-')
axes[1, 0].set_title('Tangent Wave')

# Bottom-right plot
# Add a small constant to avoid log(0)
y_log = np.log(x + 1e-9)
axes[1, 1].plot(x, y_log, 'k:')
axes[1, 1].set_title('Log Wave')

# Add a title to the entire figure
fig.suptitle('Grid of Plots')

# Adjust layout to prevent titles from overlapping
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
