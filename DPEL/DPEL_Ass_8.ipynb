```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

### Numpy Linear Spacing

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
```

### Subplots

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(x, y, 'b-')
```

## Object Oriented Programming with Matplot

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'r--')
axes.set_title('Sine Wave')
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
```

### Nested Plots

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x, y, "--b")
axes1.set_xlabel("X-axis")
axes1.set_ylabel("Y-axis")
axes1.set_title("Sine Wave")

axes2.plot(x, y, "r-")
axes2.set_xlabel("X-axis")
axes2.set_ylabel("Y-axis")
axes2.set_title("Cosine Wave")
```

### Creating Subplot

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for i in axes:
    i.set_xlabel("X-axis")
    i.set_ylabel("Y-axis")
    i.set_title("Sine Wave")
    i.plot(x, y, 'b--')
```
