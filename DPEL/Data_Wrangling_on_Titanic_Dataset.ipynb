```python
age_mean_value = df['Age'].mean()
df.fillna({'Age': age_mean_value}, inplace=True)
```

```python
df["Embarked"].ffill(inplace = True)
```

```python
df["Title"] = df['Name'].str.extract(r'([A-Za-z]+)\.', expand = True)
df['Title'].value_counts()
```

```python
mapping = {'Mlle':'Miss', "Major":"Mr", "Col":"Mr", "Sir":"Mr", "Don":"Mr", "Mme":"Miss", "Jonkheer":"Mr", "Lady":"Mrs", "Capt":"Mr", "Countess":"Mrs", "Ms":"Miss", "Dona":"Mrs"}
df.replace({'Title': mapping}, inplace = True)
```

```python
df['Title'].value_counts()
```

```python
median_age = df.groupby('Title').Age.transform('median')

fig = px.bar(df, x="Title" , y=median_age)
fig.show()
```

```python
filtered_age = df[(df.Age>40) & (df.Gender=='female') & (df.Survived==1)]
filtered_age
```
