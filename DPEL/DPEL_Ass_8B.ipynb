```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

## Histogram

```python
sns.histplot(tips["total_bill"], kde=True)
```

## Distribution

```python
sns.kdeplot(tips["total_bill"], fill=True)
```

## Jointplot

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde", hue="sex")
```

### Average tips from Male and Female

```python
avgmale = tips[tips["sex"] == "Male"]["tip"].mean()
avgfemale = tips[tips["sex"] == "Female"]["tip"].mean()
avg = tips["tip"].mean()

print(f"Average bill for Male: {avgmale}")
print(f"Average bill for Female: {avgfemale}")
print(f"Average bill overall: {avg}")
```

## Barplot

```python
sns.barplot(x="sex", y="tip", data=tips, estimator=np.std)
plt.xlabel("SEX")
plt.ylabel("TIP")
```

## Boxplot

```python
sns.boxplot(x="day", y="total_bill", data=tips, whis=1.5, hue='sex')
```

## Violinplot

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex")
```

## Swarmplot

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex")
```

## Multiple Plot Grid

```python
g = sns.PairGrid(tips, hue="sex")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```
